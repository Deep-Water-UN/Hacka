```pyspark3
import geomesa_pyspark
conf = geomesa_pyspark.configure(
    jars=['/usr/lib/spark/jars/geomesa-hbase-spark-runtime_2.11-2.1.0-m.2.jar'],
    packages=['geomesa_pyspark','pytz'],
    spark_home='/usr/lib/spark/').\
    setAppName('DeepWater-AIS-Extract')

conf.get('spark.master')

# u'yarn'

from pyspark.sql import SparkSession
spark = (
    SparkSession
    .builder
    .config(conf=conf)
    .getOrCreate()
)
```

```pyspark3
params = {
    "hbase.zookeepers": "hbase.optix-ons-local:2181",
    "hbase.catalog": "ons-historical"
}
feature = "orbcomm"
adsbx = (spark
    .read
    .format("geomesa")
    .options(**params)
    .option("geomesa.feature", feature)
    .load()
)

adsbx.createOrReplaceTempView("ais")
```

```pyspark3
df = spark.sql("""
    select
        dest,
        count(__fid__)
    from
        ais
    group by
        dest
    limit 10
""").show()
```

```pyspark3
df.write.csv("s3://optix.ons.jupyter/jupyter/dwater/RE/")
```
